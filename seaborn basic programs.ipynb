```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
import numpy as np
```

# Line Plot

```python
tip_df = sns.load_dataset('tips')
tip_df
```

```python
sns.lineplot(x='total_bill',y='tip',data=tip_df,marker='o')
```

```python
sns.lineplot(x='tip',y='size',data=tip_df,marker='o',color='g')
```

# Heatmap

```python
sns.heatmap(tip_df.isnull())
```

```python
marks = pd.DataFrame({'py':[50,55,40,35],'ML':[70,80,40,60],'AI':[30,50,81,60]})

sns.heatmap(marks,annot=True,linewidths=2,linecolor='k')
```

```python
sns.heatmap(marks,annot=True,linewidths=2,linecolor='k',cmap='tab20')
```

```python
sns.heatmap(marks,annot=True,linewidths=2,linecolor='k',cmap='tab20',cbar=False,xticklabels=False,yticklabels=False)
```

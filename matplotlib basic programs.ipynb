```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Line Plot

```python
temp = np.random.randint(30,50,15)
days = np.arange(1,16)
print(temp)
print(days)
```

```python
plt.plot(days,temp)
plt.show()
```

```python
# Parameters of plot() function
plt.plot(days,temp,color='r',marker="o",label="Temprature")
plt.title("Temprature Line Plot")
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.legend()
plt.show()
```

```python
temp_mumbai = np.random.randint(25,45,15)
temp_delhi = np.random.randint(30,45,15)
days = np.arange(1,16)
```

```python
plt.plot(days,temp_mumbai,"yo--",label="Mumbai Temp")
plt.plot(days,temp_delhi,'k*:',label="Delhi Temp")

plt.title("Delhi and Mumbai Temp")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.legend()
plt.show()
```

# Histogram Plotting

```python
stu_age = np.random.randint(15,35,100)
print(stu_age)
```

```python
plt.hist(stu_age,rwidth=0.5,color='k',label='Students Age')
plt.xlabel("Age Catagory")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

```python
stu_py = np.random.randint(15,30,100)
stu_ML = np.random.randint(15,35,100)
print(stu_py)
print(stu_ML)
```

```python
plt.hist([stu_py,stu_ML],label=['Python Students age','ML Students age'])
plt.xlabel('Students Age Catagory')
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

# Bar Plot

```python
subject = ['Python','ML','AI','Java','Django']
stu = [20,30,40,25,35]
```

```python
plt.bar(subject,stu,label="No. of Students",color='g')
plt.xlabel("Subjects")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

# Pie Plot

```python
sub = ['Python','ML','Django','AI']
marks = [40,50,60,80]
```

```python
wedgeprops = {'linewidth':5,'edgecolor':'k'}
plt.pie(marks,labels=sub,autopct="%0.01f%%",explode=[0,0,0.2,0],shadow=True,wedgeprops=wedgeprops)
plt.show()
```

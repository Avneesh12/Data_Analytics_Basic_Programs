```python
import numpy as np
import pandas as ps
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.sample()
```

```python
plt.figure(figsize=(12,9))
plt.subplot(2,2,1)
sns.boxplot(data=df['sepal_width'])
plt.subplot(2,2,2)
sns.boxplot(data=df['sepal_length'])
plt.subplot(2,2,3)
sns.boxplot(data=df['petal_length'])
plt.subplot(2,2,4)
sns.boxplot(data=df['petal_width'])
```

```python
plt.figure(figsize=(12,9))
plt.subplot(2,2,1)
sns.distplot(df['sepal_width'])
plt.subplot(2,2,2)
sns.distplot(df['sepal_length'])
plt.subplot(2,2,3)
sns.distplot(df['petal_length'])
plt.subplot(2,2,4)
sns.distplot(df['petal_width'])
```

```python
# Max
df['sepal_width'].max()
```

```python
# MIn
df['sepal_width'].min()
```

```python
# Mean
df['sepal_width'].mean()
```

```python
# Standerd deviation
df['sepal_width'].std()
```

```python
# Calculate Lower Bound
lower_bound = df['sepal_width'].mean() - 3*df['sepal_width'].std()
```

```python
lower_bound
```

```python
# Calculate Uper Bound
uper_bound = df['sepal_width'].mean() + 3*df['sepal_width'].std()
```

```python
uper_bound
```

```python
# Finding Outliers
df_out = df[(df['sepal_width']<lower_bound) | (df['sepal_width']>uper_bound)]
```

```python
df_out
```

```python
# Z-Score => Standard Score

z_score = df['z_score_sepal_width'] = df['sepal_width'].mean()/df['sepal_width'].std()
```
